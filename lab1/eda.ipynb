Можно заметить, что в файле MOCK_DATA.csv всего 1000 строк и 1000 уникальных значений в столбце id, sale_customer_id, sale_seller_id, sale_product.

```python
for i in range(0, 10):
    number = '' if i == 0 else f' ({i})'
    df = pd.read_csv(f'data/MOCK_DATA{number}.csv')
    print(f'MOCK_DATA{number}.csv')
    print(df.filter(like="id").nunique())
```

Проверим, что id=sale_customer_id=sale_seller_id=sale_product_id в каждом файлe

```python
for i in range(0, 10):
    number = '' if i == 0 else f' ({i})'
    df = pd.read_csv(f'data/MOCK_DATA{number}.csv')
    print(f'MOCK_DATA{number}.csv')
    print(df.where((df.id == df.sale_customer_id) & (df.sale_customer_id == df.sale_seller_id) & (df.sale_seller_id == df.sale_product_id)).filter(like='id').count())
```
